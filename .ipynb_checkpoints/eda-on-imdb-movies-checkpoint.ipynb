# Import Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
billion = 1000000000
million = 1000000
```

# Exploratory Data Analysis

```python
df = pd.read_csv("imdb_movies.csv")
print("Number of movies is: ", df.shape[0])
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
df.country.unique()
```

```python
status_counts = df['status'].value_counts()

plt.figure(figsize=(20,12))

plt.bar(status_counts.index, status_counts.values)
plt.title('Distribution of Movies by Status', fontsize=18)
plt.xticks(rotation=45, fontsize=18)
plt.tight_layout()
plt.show()
```

```python
country_counts = df['country'].value_counts()

plt.figure(figsize=(20,12))

plt.bar(country_counts.index, country_counts.values)
plt.title('Distribution of Movies by Country', fontsize=18)
plt.xticks(rotation=75, fontsize=16)
plt.show()
```

```python
genres = df['genre'].str.split(',')
genres_count = {}
for genres in genres:
    for genre in genres:
        genre = genre.strip()
        if genre in genres_count:
            genres_count[genre] += 1
        elif genre not in genres_count:
            genres_count[genre] = 1
            
genres_count
```

```python
keys = list(genres_count.keys())
values = list(genres_count.values())

plt.bar(keys, values)
plt.title('Distribution of Genres')
plt.xlabel('Genre')
plt.ylabel('Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
df.describe().T.style.format('{:,.2f}')
```

```python
corr = df[["score", "revenue", "budget_x"]].corr()
sns.heatmap(corr, annot=True)
```

##### From this heatmap we can know that there is a positive correlation between Revenue and Budget 
###### that we know if we increase the budget of the movie it could increase the revenue

```python
revenue_by_country = df.loc[:, ['country', 'revenue']]
revenue_by_country = revenue_by_country.groupby('country').sum()
revenue_by_country = revenue_by_country.sort_values(by="revenue", ascending=False)

plt.figure(figsize=(20, 12))
plt.bar(revenue_by_country.index, revenue_by_country['revenue']/billion)

plt.xticks(rotation=90, fontsize=16)

plt.title("Countries' Revenue From Movies")
plt.xlabel('Country', fontsize=16)
plt.ylabel('Revenue (in billions)', fontsize=16)

plt.show()
```

```python
top_10_revenues = df.sort_values("revenue", ascending=False)[:10]
plt.bar(top_10_revenues['names'], top_10_revenues['revenue']/billion)

plt.xticks(rotation=90)
plt.title("Top 10 Movies' Revenue")
plt.xlabel('Movie Title')
plt.ylabel('Revenue in Billion')

plt.show()
```

```python
df['date_x'] = pd.to_datetime(df['date_x'])
df['year'] = df['date_x'].dt.year
mean_by_year = df.groupby('year')[['revenue', 'budget_x']].mean()
```

```python
plt.figure(figsize=(30,15))
plt.bar(mean_by_year.index, mean_by_year.revenue/million, width=1)
plt.xticks(fontsize=18)
plt.xlabel("Year", fontsize=18)
plt.ylabel("Revenue (in millions)", fontsize=20)
plt.title("Revenue over Years (1903-2023)", fontsize=25)
plt.show()
```

```python
revenue_by_lang = df.loc[:, ['orig_lang', 'revenue']]
revenue_by_lang = revenue_by_lang.groupby('orig_lang').sum()
top_5_langs = revenue_by_lang.nlargest(5, 'revenue')

plt.bar(top_5_langs.index, top_5_langs['revenue']/billion)

plt.xticks(rotation=90)

plt.xlabel('Country')
plt.ylabel('Revenue (in billions)')
plt.title("Top 5 languages' Revenue")

plt.show()
```

```python
df["rev_bug"] = df["revenue"] - df["budget_x"]
top_10 = df.sort_values(by="rev_bug", ascending=True)[-10:][["names", "budget_x", "revenue", "rev_bug"]]
```

```python
plt.figure(figsize=(20,10))

df_rev_bug = pd.melt(top_10, id_vars=["names"], var_name="variable", value_vars=["budget_x", "revenue", "rev_bug"])

sns.barplot(data=df_rev_bug, x="names", y="value", hue="variable")

def format_title(title):
    return '\n'.join(title.split()[:6])

plt.xticks(rotation=45, fontsize=15)
plt.gca().set_xticklabels([format_title(t.get_text()) for t in plt.gca().get_xticklabels()])

plt.show()
```

```python
plt.scatter(df['budget_x']/billion, df['revenue']/billion)
plt.title('Budget vs Revenue')
plt.xlabel('Budget in Billion')
plt.ylabel('Revenue in Billion')
plt.show()
```
